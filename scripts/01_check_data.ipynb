```python
import re

def get_ac_no(c):
    m = re.match('.*?LHID=(\d*).*', c)
    if m:
        return int(m.group(1))
    else:
        return 0

rdf = None

for i, df in enumerate(pd.read_csv('rights.csv.gz', chunksize=500000, usecols=['url', 'khatadari'])):
    df.loc[df.khatadari=='Label', 'khatadari'] = df.loc[df.khatadari=='Label', 'url'].apply(lambda c: get_ac_no(c))
    if rdf is None:
        rdf = df[['khatadari']].copy()
    else:
        rdf = pd.concat([rdf, df[['khatadari']].copy()])
    print(i)
```
